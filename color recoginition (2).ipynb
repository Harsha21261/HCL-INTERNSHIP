```python
# Python code for Multiple Color Detection

import numpy as np
import cv2


webcam = cv2.VideoCapture(0)


while(1):

    _,imageFrame = webcam.read()


    hsvFrame = cv2.cvtColor(imageFrame, cv2.COLOR_BGR2HSV)


    red_lower = np.array([136, 87, 111], np.uint8)
    red_upper = np.array([180, 255, 255], np.uint8)
    red_mask = cv2.inRange(hsvFrame, red_lower, red_upper)


    green_lower = np.array([25, 52, 72], np.uint8)
    green_upper = np.array([102, 255, 255], np.uint8)
    green_mask = cv2.inRange(hsvFrame, green_lower, green_upper)


    blue_lower = np.array([94, 80, 2], np.uint8)
    blue_upper = np.array([120, 255, 255], np.uint8)
    blue_mask = cv2.inRange(hsvFrame, blue_lower, blue_upper)

    lower_black = np.array([0, 0, 0])
    upper_black = np.array([179, 50, 50])
    black_mask = cv2.inRange(hsvFrame, lower_black, upper_black)
    
    lower_white = np.array([0, 0, 200])
    upper_white = np.array([180, 30, 255])
    white_mask=cv2.inRange(hsvFrame, lower_white, upper_white)
    
    lower_yellow = np.array([20, 100, 100])
    upper_yellow = np.array([30, 255, 255])
    yellow_mask = cv2.inRange(hsvFrame, lower_yellow, upper_yellow)
    
    
    lower_brown = np.array([8, 60, 20])
    upper_brown = np.array([20, 255, 255])
    brown_mask = cv2.inRange(hsvFrame, lower_brown, upper_brown)
    
    lower_orange=np.array([0,50,50])
    upper_orange=np.array([30,255,255])
    orange_mask=cv2.inRange(hsvFrame,lower_orange,upper_orange)

    kernal = np.ones((5, 5), "uint8")

    red_mask = cv2.dilate(red_mask, kernal)
    res_red = cv2.bitwise_and(imageFrame, imageFrame,mask = red_mask)

    green_mask = cv2.dilate(green_mask, kernal)
    res_green = cv2.bitwise_and(imageFrame, imageFrame,mask = green_mask)

    blue_mask = cv2.dilate(blue_mask, kernal)
    res_blue = cv2.bitwise_and(imageFrame, imageFrame,mask = blue_mask)
    
    black_mask = cv2.dilate(black_mask,kernal)
    res_black = cv2.bitwise_and(imageFrame,imageFrame,mask=black_mask)
    
    white_mask=cv2.dilate(white_mask,kernal)
    res_white=cv2.bitwise_and(imageFrame,imageFrame,mask=black_mask)
    
    yellow_mask=cv2.dilate(yellow_mask,kernal)
    res_yellow=cv2.bitwise_and(imageFrame,imageFrame,mask=black_mask)
    
    brown_mask=cv2.dilate(brown_mask,kernal)
    res_brown=cv2.bitwise_and(imageFrame, imageFrame, mask=brown_mask)
    
    orange_mask=cv2.dilate(orange_mask,kernal)
    res_orange=cv2.bitwise_and(imageFrame,imageFrame,mask=orange_mask)

    contours, hierarchy = cv2.findContours(red_mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

    for pic, contour in enumerate(contours):
        
        area = cv2.contourArea(contour)
        if(area > 300):
            x, y, w, h = cv2.boundingRect(contour)
            imageFrame = cv2.rectangle(imageFrame, (x, y),(x + w, y + h),(0, 0,255, 0), 2)

            cv2.putText(imageFrame, "Red", (x, y),
                         cv2.FONT_HERSHEY_SIMPLEX, 1.0,
                          (0,0, 255, 0))

    contours, hierarchy = cv2.findContours(green_mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

    for pic, contour in enumerate(contours):
        area = cv2.contourArea(contour)
        if(area > 300):
            x, y, w, h = cv2.boundingRect(contour)
            imageFrame = cv2.rectangle(imageFrame, (x, y),
                                    (x + w, y + h),
                                    (0,255, 0), 2)

            cv2.putText(imageFrame, "Green", (x, y),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1.0, (0, 255,0, 0))


    contours, hierarchy = cv2.findContours(blue_mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        
        area = cv2.contourArea(contour)
        if(area > 300):
            x, y, w, h = cv2.boundingRect(contour)
            imageFrame = cv2.rectangle(imageFrame, (x, y),(x + w, y + h),(255, 0, 0), 2)

            cv2.putText(imageFrame, "Blue", (x, y),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1.0, (255, 0, 0,0))

    contours, hierarchy = cv2.findContours(black_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area=cv2.contourArea(contour)
        if(area>300):
            x, y, w, h = cv2.boundingRect(contour)
            imageFrame=cv2.rectangle(imageFrame, (x, y), (x+w, y+h), (0, 0,0, 255), 2)
            cv2.putText(imageFrame,"BLACK",(x,y),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        1.0,(0,0,0,255))
    contours, hierarchy = cv2.findContours(white_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area=cv2.contourArea(contour)
        if(area>300):
            x, y, w, h = cv2.boundingRect(contour)
            imageFrame=cv2.rectangle(imageFrame, (x, y), (x+w, y+h), (255, 255, 255), 2)
            cv2.putText(imageFrame,"WHITE",(x,y),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1.0,(255, 255, 255),2)
   
    contours, hierarchy = cv2.findContours(yellow_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area = cv2.contourArea(contour)
        if(area>300):
            x, y, w, h = cv2.boundingRect(contour)
            imageFrame = cv2.rectangle(imageFrame, (x, y), (x+w, y+h), (0, 255, 255), 2)
            cv2.putText(imageFrame, "YELLOW", (x, y),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1.0, (0, 255, 255), 2)
    
    contours, hierarchy = cv2.findContours(brown_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for pic, contour in enumerate(contours):
        area = cv2.contourArea(contour)
        if(area > 300):
            x, y, w, h = cv2.boundingRect(contour)
            imageFrame = cv2.rectangle(imageFrame, (x, y), (x+w, y+h), (0, 165, 255), 2)
        cv2.putText(imageFrame, "Human BROWN", (x, y), cv2.FONT_HERSHEY_SIMPLEX,
                    1.0, (0, 165, 255), 2)
        
    contours, hierarchy = cv2.findContours(orange_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for pic, contour in enumerate(contours):
        area = cv2.contourArea(contour)
        if(area>300):
            x, y, w, h = cv2.boundingRect(contour)
            imageFrame = cv2.rectangle(imageFrame, (x, y), (x+w, y+h), (0, 165, 255), 2)
            cv2.putText(imageFrame, "ORANGE", (x, y),
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1.0, (0, 165, 255), 2)
    
    cv2.imshow("Multiple Color Detection in Real-TIme", imageFrame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
webcam.release()
cv2.destroyAllWindows()
```
